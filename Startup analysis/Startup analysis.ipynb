Начнем с обработки пропусков строковых стобцов: name, homepage_url, category_list, market, status, country_code, state_code, region и city. Заменим пропуски на значения-индикаторы.

```python
#создадим список столбцов для замены пропусков
col_list=['name', 'homepage_url', 'category_list', 'market', 'status', 
          'country_code', 'state_code', 'region', 'city']

#заменим пропуски на индикатор 'unknown'
for item in col_list:
    inv_df[item]=inv_df[item].fillna('unknown')

#выведем первые 5 строк, чтобы проверить, выплнилась ли замена
display(inv_df[col_list].head())
```

Приведем содержимое строковых столбцов к единому стилю, чтобы избежать неявных дубликатов.

Столбец homepage_url преобразуем отдельно,так как ссылки могут содержать запятые и заглавные буквы. Остальные столбцы приведем к стилю snake_case. В столбце category_list оставим разделитель |.

```python
#функция преобразования содержимого столбца
def new_format(val):

    #убираем запятые
    val=val.replace(',', '')
    
    #разделяем строку пробелами
    list = val.split(' ')
    new_list=list
    
    #удаляем ненужные пробелы в начале и в конце строки
    while new_list[0]=='':
        new_list.pop(0)
        
    while new_list[-1]=='':
        new_list.pop(-1)
        
    #соединяем список с помощью разделителя '_'    
    return '_'.join(new_list)

#функция преобразования содержимого homepage_url
def new_url_format(val):
    
    #разделяем строку пробелами
    list = val.split(' ')
    new_list=list
    
    #удаляем ненужные пробелы в начале и в конце строки
    while new_list[0]=='':
        new_list.pop(0)
        
    while new_list[-1]=='':
        new_list.pop(-1)
        
    return ''.join(new_list)

#для каждого столбца из списка
for item in col_list:
    
    #для столбца homepage_url
    if item=='homepage_url':
        inv_df[item]= inv_df[item].apply(new_url_format)
        
    #для остальных строковых столбцов
    else:
        inv_df[item]= inv_df[item].apply(new_format)
        inv_df[item]=inv_df[item].str.lower()
        
#выведем первые 5 строк, чтобы проверить формат данных в столбцах
display(inv_df[col_list].head())
```

Пропуски в стоковых столбцах обработаны, значения приведены к единому стилю. Теперь обработаем полные дубликаты в данных и пропуски в funding_total_usd.

```python
#проверяем наличие явных дубликатов
print(inv_df.duplicated().sum())
```

```python
#избавимся от дубликатов 
inv_df=inv_df.drop_duplicates()
```

### 2.1. Группы по срокам финансирования
<a class="anchor" id="#21-bullet"></a>

Разделите все компании на группы в зависимости от срока финансирования:
- единичное финансирование — был всего один раунд финансирования;
- срок финансирования до года — между первым и последним раундом финансирования прошло не более года;
- срок финансирования более года.

```python
#категоризуем данные по сроку финансирования

#создаем функцию для категоризации данных 
def fund_categorize(row):
    #задаем интервал для сравнения
    interval=pd.to_timedelta('356 days')
    
    #всего один раунд финансирования
    if row['funding_rounds']==1:
        return 'единичное'
    
    elif row['funding_rounds']>1:
    
        #между первым и последним раундом финансирования прошло не более года
        if (row['last_funding_at']-row['first_funding_at'])<=interval:
            return 'срок до года'
        
        #срок финансирования более года
        elif (row['last_funding_at']-row['first_funding_at'])>interval:
            return 'срок более года'
        
        #не попадает ни под одну из категорий
        else:
            return 'нет категории'
    
    #не попадает ни под одну из категорий
    else:
        return 'нет категории'

#создаем новый столбец с категориями
inv_df['fund_cat']=inv_df.apply(fund_categorize, axis=1)

#выводим первые 5 строк с категорией
display(inv_df[['name', 'fund_cat']].head())
```

Исследуем компании с аномальным объёмом общего финансирования с помощью метода IQR отдельно по каждому сегменту.

```python
#столбец с принадлежностью к выбросам
inv_df['anomaly']=False

#проходимся по каждому сегменту и строкам, относящимся к текущему типу сегмента
for market, rows in inv_df.groupby('market'):

    #определяем границы допустимых значений для сегмента
    Q1=rows['funding_total_usd'].quantile(0.25)
    Q3=rows['funding_total_usd'].quantile(0.75)
    IQR=Q3-Q1
    lower_bound = Q1-1.5*IQR
    upper_bound = Q3+1.5*IQR
    
    #если значение объема привлеченных инвестиций уходит за границы допустимых
    #то пометим его как выброс, то есть значение будет равно true
    anomaly_series=(rows['funding_total_usd']<lower_bound)|(rows['funding_total_usd']>upper_bound) 
    
    #обновим столбец с принадлежностью к выбросам
    inv_df.loc[rows.index, 'anomaly']=anomaly_series
    
#проверяем, что новый столбец заполнился
display(inv_df[['name', 'market', 'anomaly']].head(10))    
```

В датафрейме был создан столбец, отображающий, является ли значение объема инвестиций аномальным для сегмента, к которому компания относится. Теперь определим сегменты рынка с наибольшей долей компаний, получивших аномальное финансирование.

```python
#список данных для каждого типа сегмента
anomaly_list=[]

#проходимся по каждому сегменту и строкам, относящимся к текущему типу сегмента
for market, rows in inv_df.groupby('market'):
    
    #определяем количество стартапов, соответствующих каждому из сегментов
    startup_cnt=rows.shape[0]
    
    #определяем количество аномальных значений инвестиций в сегменте и их процент, 
    #относительно количества компаний в сегменте
    anomaly_cnt=rows['anomaly'].sum()
    anomaly_per=round(rows['anomaly'].sum()/startup_cnt*100,2)

    anomaly_list.append({'market': market,
                         'startup_cnt': startup_cnt,
                         'anomaly_cnt': anomaly_cnt,
                         'anomaly_per': anomaly_per})
    
#составляем датафрейм с данными для каждого типа сегмента
anomaly_df=pd.DataFrame(anomaly_list)

#сортируем полученный датафрейм для выделения топа сегментов
top_anomaly_df=anomaly_df.sort_values(by='anomaly_per', ascending=False)

#выделяем топ-10 сегментов рынка с наибольшей долей компаний, получивших аномальное финансирование
top_anomaly_df=top_anomaly_df.head(10)

print('Топ-10 сегментов рынка по количеству стартапов с аномальными инвестициями')
display(top_anomaly_df)
print('market - сегмент рынка,')
print('startup_cnt - количество стартапов в сегменте,')
print('anomaly_cnt - количество стартапов с аномальными значениями объема инвестиций,')
print('anomaly_per - процент стартапов с аномальными значениями объема инвестиций.')
```

---
В результат анализа выбросов в каждом из сегментов рынка:
- были построены графики распределение общего объёма привлечённых инвестиций для стартапов, из которых мы определили, что распределение несимметричное с положительной скошенностью;
- определены значения финансирования типичные для стартапов, они определяются в диапозоне от 350000 до 10000000＄;
- определена минимальная сумма инвестиций, которая составила 1＄;
- определена максимальная сумма инвестиций, которая составила 30079503000＄;
- определено, что наиболее часто встречаются стартапы с общим объёмом привлечённых инвестиций в размере 2000000＄ (медиана);
- для каждого сегмента рынка с помощью метода IQR были выделены компании с аномальными значениями объема инвестиций, а также выделен топ-10 сегментов с наибольшим содержанием стартапов с аномальными инвестициями. Среди них наибольшее количество аномалий отмечается в сегментах real_estate (17.2%), entertainment	(16.67%) и consulting (16.62%).

## 4. Анализ динамики.
<a class="anchor" id="#4-bullet"></a>
### 4.1 Динамика предоставления финансирования по годам.
<a class="anchor" id="#41-bullet"></a>

Рассчитаем для каждой компании средний объём одного раунда финансирования.

```python
#создаем новый столбец и заполняем его 
#так как funding_total_usd и funding_rounds не содержат пропусков, нам нужно
#только исключить деление на 0
def mid_fund_round(row):
    
    try:
        val=round(row['funding_total_usd']/row['funding_rounds'], 2)
        
    except ZeroDivisionError:
        val=np.nan
        
    return val

new_inv_df['mid_fund_round']=new_inv_df.apply(mid_fund_round, axis=1)

#выводим первые 5 строк
display(new_inv_df[['name', 'mid_fund_round']].head())
```

### 4.2 Динамика размера общего финансирования по массовым сегментам рынка для растущих в 2014 году сегментов.
<a class="anchor" id="#42-bullet"></a>

Составим сводную таблицу, в которой указывается суммарный размер общего финансирования funding_total_usd по годам и сегментам рынка. 

```python
#строим сводную таблицу
table=pd.pivot_table(new_inv_df,
                    index='mid_funding_at',
                    columns='market',
                    values='funding_total_usd',
                    aggfunc='sum')
display(table)
```

Теперь отберем из неё только те сегменты, которые показывали рост размера суммарного финансирования в 2014 году по сравнению с 2013, и для удобства представим данные в млн. долларов.

```python
#оформляем результаты из таблицы в датафрейм
table_df=pd.DataFrame(table.to_records())
table_df.set_index('mid_funding_at', inplace=True)

#список подходящих под условие сегментов
markets=[]

#фильтруем сегменты
for item in table_df.columns:
    if (table_df[item][2014.0]>table_df[item][2013.0]) and (item not in ['mid', 'niche']):
        markets.append(item)
        
#составляем датафрейм и выводим результат, представив данные в млн. долларов
filtered_table_df=table_df[markets]
filtered_table_df=filtered_table_df/1000000
display(filtered_table_df)
```

---
Визуализируем данные с помощью линейного графика.
На графике отразите, как менялся суммарный размер общего финансирования в каждом из отобранных сегментов по годам, за которые у вас достаточно данных. Рассматривайте только массовые сегменты, а средние и нишевые исключите.

```python
#строим линейный график
plt.figure(figsize=(16, 7))

#отбираем значения, удаляя пропуски в линиях
for item in markets:    
    years=filtered_table_df.index
    line=filtered_table_df[item][years] 
    line.dropna(inplace=True)
    
    #для каждого сегмента рынка строим тренд
    plt.plot(line.index, line.values, marker='o', label=item)
    
plt.title('Динамика размера общего финансирования по массовым сегментам рынка для растущих в 2014 году сегментов.')
plt.xlabel('Год')
plt.ylabel('Общий объём привлечённых инвестиций в сегменте в млн. долларов')
plt.legend(title='Сегмент рынка', frameon=True, loc='best')
plt.grid()

plt.show()
```

---
По данным графика можно сказать, что в сегменте apps наблюдается наиболее уверенный рост общего финансирования с резким скачком за 2013-2014 год. В сегментах manufacturing и technology финансирование изменялось скачкообразно, на графике четко просматриваются пики и провалы в трендах. В остальных сегментах наблюдается рост общего финансирования, но он не такой выраженный, как у сегментов, описанных ранее.

### 4.3 Годовая динамика доли возвращённых средств по типам финансирования.
<a class="anchor" id="#43-bullet"></a>

Для каждого года и каждого вида финансирования рассчитаем нормированные значения возврата средств: то есть какую долю возвращённые средства составляют от предоставленных.

```python
#строим сводную таблицу с суммами предоставленных типом финансирования средств по годам
table_sum=pd.pivot_table(new_inv_df,
                    index='mid_funding_at',
                    values=['seed', 'venture', 'equity_crowdfunding', 
                            'undisclosed', 'convertible_note', 'debt_financing', 
                            'angel', 'grant', 'private_equity', 'post_ipo_equity', 
                            'post_ipo_debt', 'secondary_market', 'product_crowdfunding'],
                    aggfunc='sum')
display(table_sum)
```

Так как значения велики, для удобства представим данные в млн. долларов, а затем рассчитаем доли.

```python
#переводим данные таблицы в млн. долларов
table_sum_df=pd.DataFrame(table_sum.to_records())
table_sum_df.set_index('mid_funding_at', inplace=True)

#заполняем датафрейм новыми значениями и выводим результат
part_ret_df=ret_df/(table_sum_df+ 1e-60)
display(part_ret_df)
```

Мы рассчитали доли возвратов, а теперь выберем данные только по venture, debt_financing, private_equity, seed и angel.

```python
part_ret_df=part_ret_df[['venture', 'debt_financing', 'private_equity', 'seed', 'angel']]
part_ret_df=(part_ret_df-part_ret_df.min()) / (part_ret_df.max()-part_ret_df.min())
display(part_ret_df)
```

Визуализируем результат с помощью линейного графика.

```python
#строим линейный график
plt.figure(figsize=(16, 7))

#для каждого типа финансирования строим тренд
for item in part_ret_df.columns:    
    plt.plot(part_ret_df.index, part_ret_df[item], marker='o', label=item)
    
plt.title('Годовая динамика доли возвращённых средств по типам финансирования.')
plt.xlabel('Год')
plt.ylabel('Доля возвращённых средств')
plt.legend(title='Тип финансирования', frameon=True, loc='best')
plt.grid()

plt.show()
```

- Для долгового финансирования (debt_financing) доля возвращенных средств менялась непостоянно и скачками: то резко увеличиваясь, то резко уменьшаясь. С 2009 года наблюдался плавный рост практически с выходом на плато, небольшим спадом в 2013 и резким скачком вверх за 2013-2014 год.
- Для типа финансирования с помощью прямых вложений (private_equity) мы наблюдаем резкий скачек до 100% возврата средст за период 2002-2004 года, а в остальные годы возвратов не было совсем - 0%.
- Для типа финансирования с помощью посевных инвестиций (seed) доля возвращенных средств менялась непостоянно, начиная в 2000 году с большого количества возвратов (примерно 90-100% в 2000-2001 годах), за чем последовал спад и новый пик в 2006 году (чуть больше 80%) и пик в 2009 (чуть более 40%). В целом видна тенденция на снижение доли возвратов. В 2013 году доля возвратов уменьшилась до 0%. За 2013-2014 года появилась тенденция на повышение доли возвратов.
- Для финансирования с помощью бизнес-ангелов (angel) также наблюдается непостоянное изменение доли возвратов. В 2000 году возвратов по этому типу не было (0%), но на следующий год случился скачек до 100%, затем скачкообразный спад с пиками в 2006 (чуть более 60%), 2007 (примерно 60%) и 2010 (чуть более 40%) годах. С 2011 года наблюдался плавный рост доли возвратов и резкий скачек вверх в 2013-2014 году.

Таким образом, у venture, debt_financing и angel, не смотря на скачкообразное изменения значений в трендах, наблюдается наиболее устойчивый рост показателя.
